```python
import seaborn as sns
import matplotlib.pyplot as plt
a=sns.regplot(x="Ticks_hours", y="Number of birds", data=data,
                 scatter_kws={"s": 80},
                 order=2, ci=None)
plt.xlim([0,3000])
plt.ylim([0,500])

plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
c=sns.regplot(x="Ticks_hours", y="Number of infected individuals", data=data,
                 scatter_kws={"s": 80},
                 order=2, ci=None)
plt.xlim([0,3000])
plt.ylim([0,500])

plt.show()
```

```python
fig, ax = plt.subplots()
ax2 = ax.twinx() #This allows the common axes (flow rate) to be shared
sns.regplot(x="Ticks_hours", y="Number of infected birds", label='Number of infected birds', data=data, order=2, ax=ax, color="darkgreen", line_kws={'linewidth':0})
sns.regplot(x="Ticks_hours", y="Number of dead birds", data=data, order=2, label="Number of dead birds", color="darkorange", line_kws={'linewidth':0},
ax=ax2)
ax.legend(loc=1)
ax2.legend(loc=5)

ax2.set_ylim(0,1000)#This is used to set the limit for efficiency. Without this being set, the curves do not line up.
fig.tight_layout()
```
